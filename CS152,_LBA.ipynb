In this assignment, we developed a basic expert system for Minerva students who want to select a restaurant in Berlin. The system makes recommendations based on the inputs from the user. The restaurant characteristics that students can select from are class (street food, casual, fine dining), price level, proximity to the residence hall, cuisine, rating, dietary restrictions, and whether the place accepts cards or not. The logic of the system is implemented in prolog, while the UI is implemented in Python. 

The program supports multiple value answers that can be specified by the user using comma separation. The system returns a list of restaurant that satisfy user's needs. 

Menus help users choose from the variety of options.

The queries are presented to the user in a natural language rather than in the form "attribute is value".


Advanced rule types:

a) Hierarchy rules

b) Properties inferred from list membership (restaurant class)

c) Object-attribute-value

```
# The code here will ask the user for input based on the askables
# It will check if the answer is known first
from pyswip.prolog import Prolog
from pyswip.easy import *

prolog = Prolog() # Global handle to interpreter

retractall = Functor("retractall")
known = Functor("known", 3)

# Define foreign functions for getting user input and writing to the screen
def write_py(X):
    print(X)
    sys.stdout.flush()
    return True

def read_py(O,A,Y, MenuList, Msg):
    menu = ", ".join([str(item) for item in MenuList])
    string = str.capitalize(str(Msg).replace("_", " ").replace("xxxx",str(A)))
    question_str = string +"? " + menu + "\n"
    if len(MenuList) > 2:
      question_str += "Choose many with comma separation.\n"
      
    Y.unify(raw_input(question_str).strip()) #change to the new arg
    print("\n")
    return True


write_py.arity = 1
read_py.arity = 5

registerForeign(read_py)
registerForeign(write_py)

prolog.consult("KB_A.pl") # open the KB
call(retractall(known))

resultcount = 0
for soln in prolog.query("restaurant(X).", maxresult=10**3):
    resultcount = resultcount + 1
    print("Your choice is " + soln['X'])

if not resultcount:
    print("Your choice could not be identified")
```
